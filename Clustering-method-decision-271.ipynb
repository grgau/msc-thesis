### 855

```python
# Ler dataset
# df = pd.read_csv('theano-to-tf/src/tf/855_codes_prediction.csv', header=None)
df = pd.read_csv('theano-to-tf/855-attentionhcare-271-codes_prediction.csv', header=None)
df = df.iloc[:,1:]
# df = pd.read_csv('theano-to-tf/lig_855_codes_prediction.csv', header=None)
```

```python
df
```

```python
range_n_clusters = range(10, 41)

sh_score_linkage = []
db_index_linkage = []

sh_score_kmeans = []
db_index_kmeans = []

sh_score_spectral = []
db_index_spectral = []

sh_score_linkage_10 = []
db_index_linkage_10 = []

sh_score_linkage_20 = []
db_index_linkage_20 = []


for n_clusters in range_n_clusters:
#     clusterer = linkage(df, method='ward', metric='euclidean')
#     labels = fcluster(clusterer, n_clusters, criterion='maxclust')
    labels = AgglomerativeClustering(n_clusters=n_clusters).fit_predict(df)
    db_index = davies_bouldin_score(df, labels)
    sh_score = silhouette_score(df, labels)
    db_index_linkage.append(db_index)
    sh_score_linkage.append(sh_score)
    print("Linkage W\tn_clusters:{}\t silhouette score:{:.5f}\t db index:{:.5f}".format(n_clusters, sh_score, db_index))

    knn_graph = kneighbors_graph(df, 10, include_self=False)
    labels = AgglomerativeClustering(n_clusters=n_clusters, connectivity=knn_graph).fit_predict(df)
    db_index = davies_bouldin_score(df, labels)
    sh_score = silhouette_score(df, labels)
    db_index_linkage_10.append(db_index)
    sh_score_linkage_10.append(sh_score)
    print("Linkage k=10 W\tn_clusters:{}\t silhouette score:{:.5f}\t db index:{:.5f}".format(n_clusters, sh_score, db_index))

    knn_graph = kneighbors_graph(df, 20, include_self=False)
    labels = AgglomerativeClustering(n_clusters=n_clusters, connectivity=knn_graph).fit_predict(df)
    db_index = davies_bouldin_score(df, labels)
    sh_score = silhouette_score(df, labels)
    db_index_linkage_20.append(db_index)
    sh_score_linkage_20.append(sh_score)
    print("Linkage k=20 W\tn_clusters:{}\t silhouette score:{:.5f}\t db index:{:.5f}".format(n_clusters, sh_score, db_index))
    
    labels = KMeans(n_clusters=n_clusters, random_state=0).fit_predict(df)
    db_index = davies_bouldin_score(df, labels)
    sh_score = silhouette_score(df, labels)
    db_index_kmeans.append(db_index)
    sh_score_kmeans.append(sh_score)
    print("K-Means\t\tn_clusters:{}\t silhouette score:{:.5f}\t db index:{:.5f}".format(n_clusters, sh_score, db_index))    
    
    clusterer = SpectralClustering(n_clusters=n_clusters, assign_labels="discretize", random_state=0).fit(df)
    labels = clusterer.labels_
    db_index = davies_bouldin_score(df, labels)
    sh_score = silhouette_score(df, labels)
    db_index_spectral.append(db_index)
    sh_score_spectral.append(sh_score)
    print("Spectral\tn_clusters:{}\t silhouette score:{:.5f}\t db index:{:.5f}\n\n".format(n_clusters, sh_score, db_index))
```

```python
x = range(10, 41)
plt.figure(figsize=(16,13))


ax1 = plt.subplot(3, 2, 1)
ax1.plot(x, sh_score_linkage, 'r--o')
ax1.grid(axis='both')
ax1.set_title("Linkage Ward Silhouette Score")
ax1.set_ylabel('Silhouette Scores')
ax1.set_xlabel('Number of Clusters')

ax2 = plt.subplot(3, 2, 2)
ax2.plot(x, db_index_linkage, 'c--o')
ax2.grid(axis='both')
ax2.set_title("Linkage Ward DB Index")
ax2.set_ylabel('DB Index')
ax2.set_xlabel('Number of Clusters')


ax3 = plt.subplot(3, 2, 3)
ax3.plot(x, sh_score_linkage_10, 'r--o')
ax3.set_title("Linkage Ward Silhouette Score (KNN - K=10)")
ax3.grid(axis='both')
ax3.set_ylabel('Silhouette Scores')
ax3.set_xlabel('Number of Clusters')

ax4 = plt.subplot(3, 2, 4)
ax4.plot(x, db_index_linkage_10, 'c--o')
ax4.grid(axis='both')
ax4.set_title("Linkage Ward DB Index (KNN - K=10)")
ax4.set_ylabel('DB Index')
ax4.set_xlabel('Number of Clusters')


ax5 = plt.subplot(3, 2, 5)
ax5.plot(x, sh_score_linkage_20, 'r--o')
ax5.grid(axis='both')
ax5.set_title("Linkage Ward Silhouette Score (KNN - K=20)")
ax5.set_ylabel('Silhouette Scores')
ax5.set_xlabel('Number of Clusters')

ax6 = plt.subplot(3, 2, 6)
ax6.plot(x, db_index_linkage_20, 'c--o')
ax6.grid(axis='both')
ax6.set_title("Linkage Ward DB Index (KNN - K=20)")
ax6.set_ylabel('DB Index')
ax6.set_xlabel('Number of Clusters')


plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()


## OBS 25 Clusters do knn=20 parece ser o melhor resultado
```

```python
distances = ['euclidean', 'cityblock', 'minkowski']
methods = ['single', 'complete', 'average', 'weighted']

fig, axes = plt.subplots(4, 3, figsize=(40, 50))
plt.title('Hierarchical Clustering Dendrograms')

col = 0
row = 0
graphs = []
dns = []

for method in methods:
    for distance in distances:
        try:
            print(method + " - " + distance)
            Z = linkage(df, method=method, metric=distance)
            dn = dendrogram(Z, ax=axes[row, col], truncate_mode='lastp', p=50, leaf_rotation=90., leaf_font_size=8., orientation="top")
            axes[row, col].set_xlabel('Sample index', fontsize=28)
            axes[row, col].set_ylabel('Distance', fontsize=28)
            axes[row, col].set_title(method.title() + " - " + distance.title(), fontsize=40)
            graphs.append(axes[row, col])
            dns.append(dn)
        except:
            axes[row, col].set_xlabel('Sample index', fontsize=28)
            axes[row, col].set_ylabel('Distance', fontsize=28)
            axes[row, col].set_title(method.title() + " - " + distance.title(), fontsize=40)
            graphs.append(axes[row, col])
            print("Invalid: " + method + " - " + distance)
        col += 1
    row += 1
    col = 0

plt.show()
```
